```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz

url = "https://raw.githubusercontent.com/bmeelnlga/bankruptcy_data/main/bankruptcy_data.csv"
bank_data = pd.read_csv(url)
bank_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
plt.figure(figsize = (5,5))
plt.title('0: Financial.Stable || 1: Financial Unstable', fontsize=10)
sns.countplot(x="Bankrupt?",data=bank_data)
plt.show()
```

```python
mat = bank_data.corr('spearman')
plt.subplots(figsize=(30, 25))
sns.heatmap(mat, vmin=-1, vmax=1, cmap=None, center=0, #annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
x = bank_data.drop(columns=['Bankrupt?'])
x.describe()
```

```python
y = bank_data['Bankrupt?']
y.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state=32)
```

```python
estimator = DecisionTreeClassifier(criterion='entropy', max_depth=10)
estimator.fit(x_train,y_train)
```

```python
estimator.score(x_test,y_test)
```

```python
from sklearn import tree
tree.plot_tree(estimator)
plt.show()
```

```python
import graphviz
features = list(x.columns)
dot_data=tree.export_graphviz(estimator,out_file=None,feature_names=features)
graph = graphviz.Source(dot_data)
graph
```
