```python
import requests

# 发起 API 请求
response = requests.get("https://motoretag.taichung.gov.tw/DataAPI/api/ParkingSpotListAPI")
data = response.json()

# 删除包含 None 值的数据点
cleaned_data = [Position for Position in data if ['Position'] is not None]

print(cleaned_data)
```

```python
import json
import ssl
import urllib.request

url = 'https://motoretag.taichung.gov.tw/DataAPI/api/ParkingSpotListAPI'

context = ssl._create_unverified_context()

with urllib.request.urlopen(url, context=context) as jsondata:
    data = json.loads(jsondata.read().decode('utf-8-sig'))

keyword = input("請輸入台中市的行政區:")


matching_parking_lots = []

for i in data:
    for parking_lot in i['ParkingLots']:
        position = parking_lot['Position']
        if parking_lot['Position'] == None:
            del parking_lot['Position']
        else:
            if keyword.lower() in position.lower():
                matching_parking_lots.append(parking_lot)

for parking_lot in matching_parking_lots:
    position = parking_lot['Position']
    total = parking_lot['TotalCar']
    ava = parking_lot['AvailableCar']
    note = parking_lot['Notes']

    print(position)
    print('總車位:', total)
    print('剩餘車位:', ava)
    print('收費標準:', note)
    print('=====================================')
```
